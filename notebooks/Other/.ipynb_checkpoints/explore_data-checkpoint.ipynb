# Load Features

There are 16 features and 9 labels. There are 16 columns (just features) in data_test and 25 columns (features and labels) in data_train.

For feature engineering, I am interested in just the feature columns. In addition, I want to include the the feature columns from both the training and test sets to ensure new features are representative of data in both the training and test datasets (i.e. bag of words model contains words present in either of the datasets).

```python
def load_features(data_train, data_test):
    """
    Return pandas dataframe data_features: data in feature columns of data_train and data_test
    
    Param pandas dataframe data_train: training data (features + labels)
    Param pandas dataframe data_test: test data (only features)
    
    Required Libraries: pandas
    """
    
    feature_columns = data_test.columns # data_test only contains features
    
    data_features = pd.concat([data_train[feature_columns], data_test])
    
    return data_features
```

I am expecting 400277 + 50064 = 450341 rows and 16 columns.

```python
data_features = load_features(data_train, data_test)
print('data_features shape:', data_features.shape)
```

# Explore Text Features

Before exploring the text features, I will combine all text features into a single string. Then I will convert all letters to lowercase.

```python
data_features.dtypes == object
```

```python
data_features['combined_text'] = (data_features
                                      .drop(columns=['FTE', 'Total'])
                                      .fillna("")
                                      .apply(lambda x: " ".join(x), axis=1)
                                      .str.lower())
data_features['combined_text'].sample(10)
```

What are the most common words that appear?

```python
# Taken from https://stackoverflow.com/questions/16645799/how-to-create-a-word-cloud-from-a-corpus-in-python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()

show_wordcloud(data_features['combined_text'])
```

```python
TOKENS_ALPHANUMERIC = '[A-Za-z0-9 ]'
import re, string
pattern = re.compile('[\W_]+')

all_words = []
i = 0
for phrase in data_features['combined_text']:
    for word in phrase.split():
        all_words.append(pattern.sub('', word))
        
all_words = [word for word in all_words if word not in stopwords]
```

```python
from collections import Counter

counts = Counter(all_words)
counts.most_common(50)
```
