```python
# Load Requirements
import pandas as pd
import zipfile
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from wordcloud import WordCloud
import re
from collections import Counter
```

# Load Data

```python
def load_data():
    """
    Return pandas dataframe data_train: training data (features + labels)
    Return pandas dataframe data_test: test data (only features)
    
    Required Libraries: zipfile, pandas
    """
    
    # Load zipped folder with data files
    resource_archive = zipfile.ZipFile('resources.zip', 'r')

    # Load testing data
    data_test = pd.read_csv(resource_archive.open('TestData.csv'), 
                            dtype={
                                'Object_Description': str, 
                                'Program_Description': str, 
                                'SubFund_Description': str, 
                                'Job_Title_Description': str, 
                                'Facility_or_Department': str,
                                'Sub_Object_Description': str, 
                                'Location_Description': str, 
                                'FTE': float,
                                'Function_Description': str, 
                                'Position_Extra': str, 
                                'Text_4': str, 
                                'Total': float, 
                                'Text_2': str,
                                'Text_3': str, 
                                'Fund_Description': str, 
                                'Text_1': str
                            },
                            index_col=0)

    # Load training data
    data_train = pd.read_csv(resource_archive.open('TrainingData.csv'), 
                            dtype={
                                'Object_Description': str, 
                                'Program_Description': str, 
                                'SubFund_Description': str, 
                                'Job_Title_Description': str, 
                                'Facility_or_Department': str,
                                'Sub_Object_Description': str, 
                                'Location_Description': str, 
                                'FTE': float,
                                'Function_Description': str, 
                                'Position_Extra': str, 
                                'Text_4': str, 
                                'Total': float, 
                                'Text_2': str,
                                'Text_3': str, 
                                'Fund_Description': str, 
                                'Text_1': str,
                                'Function': 'category',
                                'Object_Type': 'category',
                                'Operating_Status': 'category',
                                'Position_Type': 'category',
                                'Pre_K': 'category',
                                'Reporting': 'category',
                                'Sharing': 'category',
                                'Student_Type': 'category',
                                'Use': 'category',
                            },
                             index_col=0)
    
    return data_train, data_test
```

```python
# Load Data
data_train, data_test = load_data()
print('data_train shape:', data_train.shape)
print('data_test shape:', data_test.shape)
```

# Load Features

There are 16 features and 9 labels. There are 16 columns (just features) in data_test and 25 columns (features and labels) in data_train.

For feature engineering, I am interested in just the feature columns. In addition, I want to include the the feature columns from both the training and test sets to ensure new features are representative of data in both the training and test datasets (i.e. bag of words model contains words present in either of the datasets).

```python
def load_features(data_train, data_test):
    """
    Return pandas dataframe data_features: data in feature columns of data_train and data_test
    
    Param pandas dataframe data_train: training data (features + labels)
    Param pandas dataframe data_test: test data (only features)
    
    Required Libraries: pandas
    """
    
    feature_columns = data_test.columns # data_test only contains features
    
    data_features = pd.concat([data_train[feature_columns], data_test])
    
    return data_features
```

I am expecting 400277 + 50064 = 450341 rows and 16 columns.

```python
data_features = load_features(data_train, data_test)
print('data_features shape:', data_features.shape)
```

# Explore Text Features

Before exploring the text features, I will combine all text features into a single string. Then I will convert all letters to lowercase.

```python
data_features.dtypes == object
```

```python
data_features['combined_text'] = (data_features
                                      .drop(columns=['FTE', 'Total'])
                                      .fillna("")
                                      .apply(lambda x: " ".join(x), axis=1)
                                      .str.lower())
data_features['combined_text'].sample(10)
```

## What are the most common words that appear?

```python
all_words = ', '.join([x for x in data_features['combined_text'].values])
```

```python
# Print list of most common words
all_words_list = [word for word in re.sub(r'([^\s\w]|_)+', '', all_words).split()]

counts = Counter(all_words_list)
counts.most_common(50)
```

```python
# Taken from https://stackoverflow.com/questions/38465478/wordcloud-from-data-frame-with-frequency-python
wordcloud = WordCloud(width=1200, height=1200)
wordcloud.generate_from_frequencies(frequencies=counts)
plt.figure(1, figsize=(12, 12))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig('wordcloud.png')
plt.show()
```

Observations:
- There are many stopwords (i.e. of, and, for)
- There are multiple forms of words (i.e. instructional/instruction, employees/employee)
- When dashes were removed, some words were fused together (i.e. professionalinstruction
